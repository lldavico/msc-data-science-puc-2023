```python
import wikipedia
import wordcloud
import wordtree
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
page = wikipedia.page("Square", auto_suggest=False)
content = page.content
content
```

```python
stopwords = set(wordcloud.STOPWORDS) 
wc = wordcloud.WordCloud(background_color='white', max_words=200,
                stopwords= stopwords)
wc.generate(content)
plt.rcParams["figure.figsize"] = [6, 4]
plt.imshow(wc, interpolation='bilinear', aspect='auto')
plt.axis('off')
plt.show()
```

```python
page = wikipedia.page("Cloud", auto_suggest=False)
content = page.content
```

```python
mask = np.array(Image.open('cloud.png'))
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
stopwords = set(wordcloud.STOPWORDS) 
wc = wordcloud.WordCloud(background_color='white', mask=mask, max_words=200,
                stopwords= stopwords)
wc.generate(content)
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
plt.imshow(wc, interpolation='bilinear', aspect='auto')
plt.axis('off')
plt.show()
```

```python
page = wikipedia.page("Cat", auto_suggest=False)
content = page.content
```

```python
mask = np.array(Image.open('mochi-mask.png'))
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
plt.imshow(mask)
plt.axis('off')
plt.show()
```

```python
stopwords = set(wordcloud.STOPWORDS) 
wc = wordcloud.WordCloud(background_color='white', mask=mask, max_words=200,
                stopwords= stopwords)
wc.generate(content)
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
plt.imshow(wc, interpolation='bilinear', aspect='auto')
plt.axis('off')
plt.show()
```

```python
g = wordtree.search_and_draw(corpus = [content], keyword = "cat")
g.render()
```

```python
image = np.array(Image.open('cat.gv.png'))
plt.rcParams["figure.figsize"] = [16, 16]
plt.imshow(image)
plt.axis('off')
plt.show()
```
