##  C. Histograma

```python
# Veremos la distribución de la variable tip (propina)
# con un histograma

ax = sns.displot(tips.tip)
```

```python
sns.displot(tips.tip, kde = True, color="red", bins=25)
```

```python
sns.histplot( tips, x="tip", kde = False, hue="time")
plt.show()
```

Finalmente, podríamos querer poner estos histogramas pero en gráficos uno al lado del otro.

```python
f, axes = plt.subplots(2, 1, figsize=(12, 4), sharex=True)
sns.histplot( tips[tips.time == "Lunch"], x="tip", kde = True
             , color="red", label="Propinas en Almuerzo", ax=axes[0])
sns.histplot( tips[tips.time == "Dinner"], x="tip", kde = True
             , color="skyblue", label="Propinas en Cena", ax=axes[1])
```

```python
sns.displot(tips.tip, kind="kde" , color="red")
```

```python
sns.displot(tips, x="tip", kind="kde" , hue="time")
```

##  D. Gráfico de Caja (Boxplot)


Haremos un gráfico de caja con la información de propinas para comparar la información que nos muestra respecto a un histograma. Recuerde que la linea dentro de la caja es la mediana.

```python
# Veremos la distribución de la variable tip (propina)
# con un gráfico de cajas

ax = sns.boxplot(tips.tip)
```

```python
# para forzar que quede vertical y cambiarle el color

ax = sns.boxplot(tips.tip, orient="v", color="red")
```

```python
# ahora veamos si cambian según el día de la semana, y cambiemos
# la paleta de colores por defecto

sns.boxplot(x="day", y="tip", data=tips, palette="Set3")
```

##  E. Gráfico de Violín


Recordemos que el gráfico de violín es similar al boxplot, pero además agrega una estimación de densidad de probabilidad para visualizar mejor la distribución de datos

```python
# Distribución de las cuentas en el restaurant con gráfico de violín

ax = sns.violinplot(x=tips["total_bill"])
```

```python
# Revisemos ahora cómo cámbia por día de la semana
ax = sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
# Podemos ademas separar estos gráficos en un segundo nivel, 
# dependiendo de si la mesa es para fumadores o no

ax = sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted")
```
