# Pregunta 3
La comuna comienza en la coordenada 0 y termina en la coordenada 40, tanto para X como 
para Y. Si se divide territorialmente la comuna de la siguiente manera:

 a)  En X, desde la coordenada 0 a la 22 (inclusive) se define Oeste (O)
 b.  En X, desde la coordenada 23 a la 40 (inclusive) se define Este (E)
 c.  En Y, desde la coordenada 0 a la 27 (inclusive) se define Norte (N)
 d.  En Y, desde la coordenada 28 a la 40 (inclusive) se define Sur (S) 


Considerando esta división territorial, las calles de la comuna se clasifican como zonas: N-O, N-E, S-O, S-E. Bajo esta clasificación, ¿hay calles que no puedan ser clasificadas, y entonces, que la calle comience en una zona y termine en otra zona? 

# Respuesta

```python
espacio_comuna = [(x, y) for y in (0, 40) for x in (0, 40)]
espacio_comuna
```

Creamos la logica para encapsular una calle en una 'zona' y si esta no se encuentra en una zona, dejamos la columna como NaN.

```python
# Chequeamos que la calle este dentro del rango 'x' y rango 'y' de la zona.
def street_in_zone(comienzo, fin, zone):
    comienzo_x = int(comienzo[0])
    comienzo_y = int(comienzo[1])
    fin_x = int(fin[0])
    fin_y = int(fin[1])

    range_x = range(zone[0][0], zone[1][0] + 1)
    range_y = range(zone[0][1], zone[1][1] + 1)

    if (comienzo_x in range_x and fin_x in range_x) and (comienzo_y in range_y and fin_y in range_y):
        return True
    return False

# Definimos las zonas y sus coordenadas e iteramos por zona para chequear si la calle esta completamente adentro
def get_zona(comienzo, fin):
    N_O = ((0, 0), (22, 27))
    N_E = ((23, 0), (40, 27))
    S_O = ((0, 28), (22, 40))
    S_E = ((23, 28), (40, 40))
    zones = [N_O, N_E, S_O, S_E]
    dict_zones = dict(zip(zones, ['N-O', 'N-E', 'S-O', 'S-E']))

    comienzo = comienzo.replace('(', '').replace(')', '').split(',')
    fin = fin.replace('(', '').replace(')', '').split(',')

    for zone in zones:
        if street_in_zone(comienzo, fin, zone):
            return dict_zones[zone]
        
    return 'Sin zona'

ciudad_df['zona'] = ciudad_df.apply(lambda x: get_zona(x['comienzo'], x['fin']), axis=1)
ciudad_df.sample(10)
```

Ahora, chequeamos si hay calles con el valor 'Sin zona en la columna 'zona'. En dicho caso, esas calles no pueden ser clasificadas ya que están en más de una zona.

```python
calles_sin_zona = ciudad_df[ciudad_df['zona'] == 'Sin zona']
calles_sin_zona
```

```python
total_calles_sin_zona = len(calles_sin_zona)
print(f'Vemos que hay {total_calles_sin_zona} calles que comienzan en una zona y terminan en otra.')
```

# Pregunta 5
¿Cuál es el promedio y desviación estándar de la cantidad de kilos de basura generada por cada zona (si existen, excluya de este análisis las calles que no pueden ser clasificadas en una zona)?

# Respuesta

Filtramos solo por las calles que tienen una zona.

```python
ciudad_df_5 = ciudad_df[ciudad_df['zona'] != 'Sin zona']
ciudad_df_5
```

Agrupamos el total de kilos de basura que tiene cada calle.

```python
basura_df_5 = basura_df.groupby('calle').sum()[['kg_basura']]
basura_df_5.reset_index(inplace=True)
basura_df_5
```

Hacemos un merge entre este dataset filtrado y el dataframe con datos de total de kg de basura por calle. Importante considerar que, al hacer un merge de tipo 'inner' (un INNER JOIN), solo consideraremos la intersección de calles de ambos de dataset, por lo que no se considerarán las calles en más de una zona.

```python
basura_df_5.head()
```

```python
merged_df_5 = ciudad_df_5.merge(basura_df_5, how='inner', on='calle')
merged_df_5
```

Agrupamos la cantidad de kilos de basura por zona, calculando su promedio y luego su desviación estandar.

```python
grouped_df_5_mean = merged_df_5[['zona', 'kg_basura']].groupby(['zona']).mean()
grouped_df_5_mean.reset_index(inplace=True)
grouped_df_5_mean
```

```python
grouped_df_5_std = merged_df_5[['zona', 'kg_basura']].groupby(['zona']).std()
grouped_df_5_std.reset_index(inplace=True)
grouped_df_5_std
```

Finalmente, juntamos ambos dataset y obtenemos el promedio y desviación estándar de basura por cada zona.

```python
df_5_final = grouped_df_5_mean.merge(grouped_df_5_std, how='inner', on='zona')
df_5_final.rename(columns={'kg_basura_x': 'promedio_basura', 'kg_basura_y': 'dev_standard_basura'}, inplace=True)
df_5_final
```
