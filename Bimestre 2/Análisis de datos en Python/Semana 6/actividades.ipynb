### P3

```python
boletas = boletas.fillna('-')
boletas
```

```python
def get_price_boleta(row, productdf):    
    price_boleta = 0.0
    for p in row:
        if p != '-':
            price_p = productdf[productdf['Producto'] == p]['Precio'].values[0]
            price_boleta += price_p
    return price_boleta

boletas['Suma precios'] = boletas.apply(lambda x: get_price_boleta(x[[f'Producto {i+1}' for i in range(10)]], productos), axis=1)
boletas
```

```python
boletas[boletas['Valor Boleta'] != boletas['Suma precios']]
```

```python
boletas['Diferencia'] = boletas['Valor Boleta'] - boletas['Suma precios']
boletas[boletas['Diferencia'] > 0]
```

```python
precio = boletas['Diferencia'][97]
precio
```

```python
producto1 = []
for i in range(len(boletas)):
    if boletas['Diferencia'][i] != 0:
        precio = boletas['Diferencia'][i]
        producto = productos.loc[productos['Precio'] == precio].reset_index()['Producto'][0]
        producto1.append(producto)
    else:
        producto1.append(boletas['Producto 1'][i])

boletas['Producto 1'] = producto1
boletas[boletas['Diferencia'] > 0]
```

### P1

```python
def total_products(row):  
    total_products = 0
    for p in row:
        if p != '-':
            total_products += 1
    return total_products

boletas['Total productos'] = boletas.apply(lambda x: total_products(x[[f'Producto {i+1}' for i in range(10)]]), axis=1)
boletas
```

```python
corr = boletas.corr()
round(corr['Valor Boleta']['Total productos'], 2)
```

### P2

```python
families = productos['Familia'].unique()
families

def get_families(row, productdf, family):
    is_in_family = 0
    for p in row:
        if p != '-':
            p_family = productdf[productdf['Producto'] == p]['Familia'].values[0]
            if p_family == family:
                is_in_family = 1
    return is_in_family

for fam in families:
    print(fam)
    boletas[fam] = boletas.apply(lambda x: get_families(x[[f'Producto {i+1}' for i in range(10)]], productos, fam), axis=1)
boletas
```
