Luego incorporamos todos los resultados obtenidos del Jupyter Notebook de las preguntas formativas 11.

```python
pro = ['Producto 1','Producto 2','Producto 3','Producto 4','Producto 5',
             'Producto 6','Producto 7','Producto 8','Producto 9','Producto 10']
suma_bol = []
for i in range(1, len(boletas)+1):
    suma = 0
    for producto in pro:
        if pd.isna(boletas[producto][i]) != True:
            suma += productos.loc[productos['Producto'] == boletas[producto][i]].reset_index()['Precio'][0]
        else:
            pass
    suma_bol.append(suma)

boletas['Suma_productos'] = suma_bol
boletas['Diferencia'] = boletas['Valor Boleta'] - boletas['Suma_productos']

producto1 = []
for i in range(1, len(boletas)+1):
    if boletas['Diferencia'][i] != 0:
        precio = boletas['Diferencia'][i]
        producto = productos.loc[productos['Precio'] == precio].reset_index()['Producto'][0]
        producto1.append(producto)
    else:
        producto1.append(boletas['Producto 1'][i])

boletas['Producto 1'] = producto1

boletas
```

### Pregunta 1
Correlación entre cantidad de productos en cada boleta y precio de la boleta

```python
# Se utiliza el Dataframe generado recién, que contiene todos los productos que debe tener la boleta
boletasP1 = boletas[['Producto 1','Producto 2','Producto 3','Producto 4','Producto 5',
             'Producto 6','Producto 7','Producto 8','Producto 9','Producto 10', 'Valor Boleta']]
boletasP1
```

```python
conteo_productos = []
for i in range(1, len(boletasP1)+1):
    conteo_productos.append(10 - boletasP1.loc[[i]].isna().sum().sum())
boletasP1['Conteo_p'] = conteo_productos
boletasP1
```

```python
boletasP1['Valor Boleta'].corr(boletasP1['Conteo_p']).round(2)
```

### Pregunta 2
Generar una columna que diga si existe alguna bebida (producto familia bebida) en la boleta

```python
# 1 si hay bebida 0 si no hay
pro = ['Producto 1','Producto 2','Producto 3','Producto 4','Producto 5',
             'Producto 6','Producto 7','Producto 8','Producto 9','Producto 10']
bebida_col = []
for i in range(1, len(boletasP1)+1):
    bebida = 0
    for producto in pro:
        if pd.isna(boletasP1[producto][i]) != True:
            familia = productos.loc[productos['Producto'] == boletasP1[producto][i]].reset_index()['Familia'][0]
            if familia == 'Bebidas':
                bebida = 1
        else:
            pass
    bebida_col.append(bebida)
    

boletasP1['Bebida'] = bebida_col
boletasP1
```

```python
boletasP1[boletasP1['Bebida'] == 1]['Bebida'].sum()
```

### Pregunta 3
Generar una columna que diga si existe algun carne (producto familia carnes) en la boleta

```python
# 1 si hay bebida 0 si no hay
pro = ['Producto 1','Producto 2','Producto 3','Producto 4','Producto 5',
             'Producto 6','Producto 7','Producto 8','Producto 9','Producto 10']
carne_col = []
for i in range(1, len(boletasP1)+1):
    carne = 0
    for producto in pro:
        if pd.isna(boletasP1[producto][i]) != True:
            familia = productos.loc[productos['Producto'] == boletasP1[producto][i]].reset_index()['Familia'][0]
            if familia == 'Carnes':
                carne = 1
        else:
            pass
    carne_col.append(carne)
    

boletasP1['Carne'] = carne_col
boletasP1
```

```python
boletasP1[boletasP1['Carne'] == 1]['Carne'].sum()
```

### Pregunta 4

Tabla de frecuencia entre "Carne" y "Bebida"
Hacer una tabla en que las filas son con/sin carne y las columnas con/sin bebida (4 lugares en la matriz). Colocar en cada posición la cuenta de la cantidad boletas que cae en cada clasificación (sin carne sin bebida; sin carne con bebida; con carne sin bebida; con carne con bebida)

```python
nada = 0
bebida = 0
carne = 0
todo = 0
for i in range(1, len(boletas)+1):
    beb = boletasP1['Bebida'][i]
    car = boletasP1['Carne'][i]
    if beb == 0 and car == 0:
        nada += 1
    elif beb == 1 and car == 0:
        bebida += 1
    elif beb == 0 and car == 1:
        carne += 1
    else:
        todo +=1

columnas = {'Sin_Bebida': [nada,carne], 'Con_Bebida': [bebida,todo]}
cuenta = pd.DataFrame(data = columnas, index = ['Sin_Carne', 'Con_Carne'])
cuenta
```

### Pregunta 5

Tabla de frecuencia entre "Carne" y "Bebida"
Hacer una tabla en que las filas son con/sin carne y las columnas con/sin bebida (4 lugares en la matriz). Colocar en cada posición el promedio del valor de las boletas en cada clasificación (sin carne sin bebida; sin carne con bebida; con carne sin bebida; con carne con bebida)

```python
nada = []
bebida = []
carne = []
todo = []
for i in range(1, len(boletasP1)+1):
    beb = boletasP1['Bebida'][i]
    car = boletasP1['Carne'][i]
    if beb == 0 and car == 0:
        nada.append(boletasP1['Valor Boleta'][i])
    elif beb == 1 and car == 0:
        bebida.append(boletasP1['Valor Boleta'][i])
    elif beb == 0 and car == 1:
        carne.append(boletasP1['Valor Boleta'][i])
    else:
        todo.append(boletasP1['Valor Boleta'][i])

columnas = {'Sin_Bebida': [np.mean(nada),np.mean(carne)], 'Con_Bebida': [np.mean(bebida),np.mean(todo)]}
prom = pd.DataFrame(data = columnas, index = ['Sin_Carne', 'Con_Carne'])
prom
```
