##Pregunta 2

Con tu esquema en mano, debes escribir los comandos SQL para crear las tablas en tu esquema. Esto requiere tener cinco o más comandos de tipo “CREATE TABLE…” que deben poder ejecutarse correctamente en SQL. Quiere decir, si nosotros ejecutamos estos comandos, el motor de SQL no nos dará un error. Para esto, te recomendamos correr tus comandos directamente en el siguiente campo de Notebook en Google Colab:

```python
%%sql


CREATE TABLE IF NOT EXISTS PERSONA(
  rut varchar(15) PRIMARY KEY,
  name varchar(50),
  email varchar(30),
  phone_number varchar(20)
);


CREATE TABLE IF NOT EXISTS TALLER(
  taller_name varchar(50) PRIMARY KEY,
  professor_name varchar(50),
  horario timestamp,
  price int,
  duration float
);

CREATE TABLE IF NOT EXISTS LUGAR(
  lugar_name varchar(50) PRIMARY KEY,
  street_name varchar(50),
  street_number int,
  zip_code varchar(10),
  comuna varchar(50),
  region varchar(30)
);

CREATE TABLE IF NOT EXISTS PARTICIPACION(
  rut VARCHAR(15),
  taller_name varchar(50),
  primary key(rut, taller_name)
);

CREATE TABLE IF NOT EXISTS UBICACION(
  taller_name varchar(50),
  lugar_name varchar(50),
  primary key(taller_name, lugar_name)
);

```

## Otras queries útiles

Todos las personas del 'Taller 11'

```python
%%sql

SELECT PERSONA.name as nombre, PERSONA.email as correo, TALLER.taller_name as taller_name
FROM PERSONA
INNER JOIN PARTICIPACION ON PERSONA.rut = PARTICIPACION.rut
INNER JOIN TALLER ON TALLER.taller_name = PARTICIPACION.taller_name
WHERE TALLER.taller_name = 'Taller 11';
```

Personas con rut comenzado en '3' y con taller ubicado en la region metropolitana

```python
%%sql

SELECT DISTINCT PERSONA.name as name, PERSONA.rut as rut, TALLER.taller_name as taller_name
FROM PERSONA
INNER JOIN PARTICIPACION ON PERSONA.rut = PARTICIPACION.rut
INNER JOIN TALLER ON PARTICIPACION.taller_name = TALLER.taller_name
INNER JOIN UBICACION ON TALLER.taller_name = UBICACION.taller_name
INNER JOIN LUGAR ON UBICACION.lugar_name = LUGAR.lugar_name
WHERE PERSONA.rut LIKE '3%' AND LUGAR.region = 'Metropolitana'
```
