Se crea una función para calcular el total de goles por partido

```python
def total_goles(string): #estamos creando la función que se llamará "total_goles", y que recibirá un string como input y retornará un entero
    #goles será un arreglo que almacena las "partes" del string que separaremos con la función split
    goles = string.split('–') #función split separa el string en 2 partes, la que está antes del "-" (escogido como separador en este caso) y la que está después
    total_goles = int(goles[0])+int(goles[1]) #sumamos goles[0] (que tiene la parte izquierda del string) con goles[1] (que tiene la parte derecha del string), transformándo los string en enteros (con la función "int")
    return total_goles
```

Se crea la columna total de goles aplicando la función creada a la columna "Score" de la dataframe.

```python
laliga['Total de goles']= laliga['Score'].apply(total_goles)
```

Se seleccionan columnas de interés para mostrar cómo se calcula el total de goles por partido.

```python
#con el comando loc seleccionamos las columnas que deseamos mostrar. Los ":" son para decir que queremos todas las filas
laliga.loc[:,['Date','Local','Score','Visitante','Total de goles']]
```

## Ejemplo 2
Crearemos dos funciones para crear las columnas de los goles realizados por los locales y visitantes para luego poder compararlo con los goles esperados.

```python
#estas dos funciones se costruyen de manera similar a la función "total_goles" recién explicada.
def golesvisita(marcador):
    goles = marcador.split('–')
    return int(goles[1])

def goleslocal(marcador):
    goles = marcador.split('–')
    return int(goles[0])

laliga['Goles Local'] = laliga['Score'].apply(goleslocal) #construimos la columna "Goles Local" y la poblamos con el resultado de aplicar la función goles local al "Score" del partido
laliga['Goles Visitante'] = laliga['Score'].apply(golesvisita) #construimos la columna "Goles Local" y la poblamos con el resultado de aplicar la función goles visita al "Score" del partido
```

Se crea la columna de comparación entre los goles realizados y esperados por cada equipo. Luego, se eligen columnas para mostrar y estudiar las nuevas columnas creadas.

```python
#similar a las funciones golesvisita y goleslocal, pero le entregaremos, además del marcador real, el marcador esperado (columnas "xG" y "xG.1")
def comp_goleslocal(marcador,xg):
    goles = marcador.split('–')
    gol_local = int(goles[0])
    if float(xg)>gol_local: #compararemos si los goles que se esperaba que marcaría el equipo local fueron menos que los que marcó
        return 1 #la función retornará un 1 si se cumple la condición
    else:
        return 0 #y retornará un 0 si no se cumple

def comp_golesvisita(marcador,xg):
    goles = marcador.split('–')
    gol_visita = int(goles[1])
    if float(xg)>gol_visita: #compararemos si los goles que se esperaba que marcaría el equipo visitante fueron menos que los que marcó
        return 1 #la función retornará un 1 si se cumple la condición
    else:
        return 0 #y retornará un 0 si no se cumple

laliga['Comparación goles local vs xg'] = laliga.apply(lambda x: comp_goleslocal(x['Score'], x['xG']),axis=1) #creamos una nueva columna aplicando a cada partido (fila del Dataframe) la función comp_goleslocal sobre el marcador real (Score) y los goles esperados del equipo local (xG)
laliga['Comparación goles visita vs xg'] = laliga.apply(lambda x: comp_golesvisita(x['Score'],x['xG.1']),axis=1)
```

```python
#mostramos las columnas de interés del Dataframe resultante
laliga.loc[:,['Local','Visitante','Score','xG','xG.1', 'Comparación goles local vs xg',
       'Comparación goles visita vs xg']]
```

## Ejercicio 2
Escriba el código para crear una variable que indique los partidos de la liga, en todos los torneos, que hayan tenido más de 20.000 espectadores.

Primero, convertiremos la variable multiplicando por 1000 para que pandas reconozca adecuadamente la magnitud de la cifra (porque en este caso la asistencia está en miles de asistentes).

```python
laliga['Attendance']= laliga['Attendance']*1000 #reemplazamos el dato actual de asistencia por el nuevo (asistencia x 1000)
```

Se crea una variable que es 1 si la asistencia al partido fue mayor a 20000 espectadores y 0 en caso de que no.

```python
#la función "lambda" nos permite hacer esto. Indicamos que colocaremos un 1 en la nueva columna "Asistencia mayor a 20k" si 
#la asistencia de ese partido es >20000, y 0 si no lo es.
laliga['Asistencia mayor a 20k'] = laliga.apply(lambda x: 1 if x['Attendance']>20000 else 0,axis = 1)
```
