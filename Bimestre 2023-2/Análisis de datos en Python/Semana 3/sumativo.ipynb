## P3 Formativo 1

```python
def goleslocal(marcador):
    goles = marcador.split('–')
    return int(goles[0])
    
def golesvisita(marcador):
    goles = marcador.split('–')
    return int(goles[1])


def get_winner(goles_local, goles_visita):
    if goles_local > goles_visita:
        return 'Local'
    elif goles_visita > goles_local:
        return 'Visitante'
    else:
        return 'Empate'
    
def two_or_more_goals_win(goles_local, goles_visita):
    if (goles_visita + 2) <= goles_local:
        return True
    elif (goles_local + 2) <= goles_visita:
        return True
    else:
        return False

laliga['Goles local'] = laliga['Score'].apply(goleslocal)
laliga['Goles visita'] = laliga['Score'].apply(golesvisita)
laliga['Ganador'] = laliga.apply(lambda x: get_winner(x['Goles local'], x['Goles visita']), axis=1)
laliga['goals>=2'] = laliga.apply(lambda x: two_or_more_goals_win(x['Goles local'], x['Goles visita']), axis=1)
laliga.head(10)
```

```python
def resultado_equipo(string, local, visita):
    goles = string.split('–')
    if int(goles[0])>int(goles[1]):
        return local
    if int(goles[0])<int(goles[1]):
        return visita
    else: 
        return 'Empate'
    
df['Ganador'] = df.apply(lambda x: resultado_equipo(x['Score'],x['Local'],x['Visitante']),axis=1)
df
```

```python
filter_p6_sum = df[(df['Torneo'] == 'Primera división') & (df['Temporada'] == '21/22')]
```

```python
def tabla_posicion(data,fecha,equipos):
    tabla = {}
    for equipo in equipos:
        data_usar = data[(data['Wk']<=fecha)&((data['Local']==equipo)|(data['Visitante']==equipo))]
        puntos = 0
        for i in range(len(data_usar)):
            ganador = str(data_usar.iloc[i]['Ganador'])
            if ganador == equipo:
                puntos += 3
            if ganador == 'Empate':
                puntos +=1
        tabla[equipo] = puntos
    df = pd.DataFrame([[key, tabla[key]] for key in tabla.keys()], columns=['Equipo', 'Puntos'])
    df = df.sort_values(by='Puntos',ascending = False)
    return df


def primera_est(data,tabla,fecha,equipo):
    dato = data[(data['Wk']>fecha)&((data['Local']==equipo)|(data['Visitante']==equipo))]
    indice = tabla[tabla['Equipo']==equipo].index.values.astype(int)[0]
    puntos = tabla.loc[indice,'Puntos']
    for i in range(len(dato)):
        if dato.iloc[i]['Local']==equipo:
            rival = dato.iloc[i]['Visitante']
            index = tabla[tabla['Equipo']==rival].index.values.astype(int)[0]
            puntos_rival = tabla.loc[index,'Puntos']
        else:
            rival = dato.iloc[i]['Local']
            index = tabla[tabla['Equipo']==rival].index.values.astype(int)[0]
            puntos_rival = tabla.loc[index,'Puntos']
        if puntos<puntos_rival:
            puntos+=3
        if puntos == puntos_rival:
            puntos +=1
    return puntos
```
