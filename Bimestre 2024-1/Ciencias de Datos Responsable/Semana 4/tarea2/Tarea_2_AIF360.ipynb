### Pregunta 1: Realice el pre-procesamiento de los datos e implemente un modelo de clasificación (Regresión Logística). Obtenga el accuracy del modelo sobre los datos de testeo (1 pto).

#### Indicaciones adicionales:

- Recuerde utilizar random_state o seeds según sea necesario para la reproducibilidad de sus resultados.

### Respuesta

```python
# Seteamos una semilla
seed = 10
```

```python
# Haremos el pre-procesamiento de los datos
display(df.info())
display(df.head())
```

```python
# Binarizamos las variables binarias en 0 y 1

df['income'] = df['income'].apply(lambda x: 1 if x == '>50K' else 0)
df['sex'] = df['sex'].apply(lambda x: 1 if x == 'Male' else 0)
```

```python
# Transformamos a 'category' las columnas 'object'

for col in df.select_dtypes(include='object').columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category')

df.info()
```

```python
# Separamos X e Y.
target = 'income'

X = df.loc[:, df.columns != target]
y = df.loc[:, df.columns == target]
```

```python
# Obtenemos variables dummies

df1 = X.copy()

ignore = ['sex', target]
for catcol in df1.select_dtypes(include='category').columns:
    if catcol in ignore:
      pass
    else:
      dummies = pd.get_dummies(X[catcol])
      dummies.columns = [catcol+ '_' + x for x in dummies.columns]
      X = pd.concat([X, dummies], axis=1).drop(columns=[catcol])

X.head()
```

```python
# Datos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=908)
```

```python
# Construimos la regresion logistica
import numpy as np
from sklearn.linear_model import LogisticRegressionCV

lr = LogisticRegressionCV(solver='liblinear', cv=10, random_state=seed)
lr.fit(X_train, np.ravel(y_train))
```

```python
# Obtenemos el accuracy sobre los datos de testeo
from sklearn.metrics import accuracy_score


y_pred_proba = lr.predict_proba(X_test)[:,1]
y_pred = y_pred_proba >= 0.5

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

### Pregunta 2: Evalúe las clasificaciones del modelo de la pregunta anterior a través de las métricas de Fairness facilitadas en AIF360 utilizando la raza o sexo como atributo protegido (elija una). Destaque alguna métrica de fairness que le sea de interés (1 pto).

#### Indicaciones adicionales:

- Obtenga tanto las intancias de BinaryLabelDatasetMetric como de ClassificationMetric y calcule al menos una métrica de Fairness.

- Considere 'Black' como grupo no privilegiado para la raza, 'White' como grupo privilegiado para la raza.
- Considere 'Female' como grupo no privilegiado para el sexo, 'Male' como grupo privilegiado para el sexo.

- Está invitado(a) a realizar el análisis para ambos atributos sensibles, aunque **no es obligatorio**.

### Respuesta

```python
# Protegeremos el atributo sexo como atributo protegido
privileged_group = [{'sex': 1}]
unprivileged_group = [{'sex': 0}]
```

```python
# Preparamos el dataset de test

df_test = X_test.copy()
df_test[target] = y_test

df_test
```

```python
# Instanciamos el dataset de test por medio de BinaryLabelDataset
from aif360.datasets import BinaryLabelDataset

attributes_params = dict(protected_attribute_names=['sex'], label_names=[target])
dt_test = BinaryLabelDataset(df=df_test, **attributes_params,
                              favorable_label=1, unfavorable_label=0)
```

```python
# Instanciamos la metrica de test por medio de BinaryLabelDatasetMetric
from aif360.metrics import BinaryLabelDatasetMetric

metric_sex_test_df = BinaryLabelDatasetMetric(dt_test,
                                        unprivileged_groups=unprivileged_group,
                                        privileged_groups=privileged_group)
```

```python
# Utilizamos las inferencias anteriores
df_pred = X_test.copy()
df_pred[target] = y_pred
```

```python
# Repetimos el proceso de calcular el dataset según los grupos privilegiados y no privilegiados

dt_pred = BinaryLabelDataset(df=df_pred, **attributes_params,
                             favorable_label=1, unfavorable_label=0)
metric_sex_pred_df = BinaryLabelDatasetMetric(dt_pred,
                                            unprivileged_groups=unprivileged_group,
                                            privileged_groups=privileged_group)
```

```python
# Utilizando ClassificationMetric, calculamos la diferencia entre las predicciones y los datos de entrenamiento
from aif360.metrics import ClassificationMetric

pred_vs_test_classification_metric = ClassificationMetric(dt_test, dt_pred,
                                                    unprivileged_groups=unprivileged_group,
                                                    privileged_groups=privileged_group)
```

```python
# Calculamos una métrica de fairness, en este caso 'Statistical Parity Difference'

# Para el dataset de test
print(f'Test Statistical Parity Difference (sex): {metric_sex_test_df.mean_difference()}')
```

Vemos que para los datos de test, tenemos que la métrica de diferencia estadística es negativa. Esto significa que la probabilidad de pertenecer a la clase positiva para el sexo femenino es menor que para el sexo masculino, lo que indica presencia de sesgo.

```python
# Replicamos lo mismo pero para las predicciones
print(f'Pred Statistical Parity Difference (sex): {metric_sex_pred_df.mean_difference()}')
```

Similar a los datos de test, tenemos que la métrica de diferencia estadística es negativa para nuestras predicciones, lo que aún indica sesgo en el modelo entrenado y en los datos, confirmado porque el modelo fue entrenado con datos sesgados, por ende las predicciones también repetirán el patrón.

```python
# Ahora comparamos el ratio de error entre los datos de test y las predicciones
print(f'Error rate difference (unprivileged error rate - privileged error rate)= {pred_vs_test_classification_metric.error_rate_difference()}\n')
```

Por otra parte, tenemos que la diferencia de la tasa de error entre ambos dataset es negativa. Esto indica que el ratio de error para el grupo no privilegiado (sexo femenino) es menor que para el grupo privilegiado (sexo masculino). En otras palabras, el modelo tiene mejores predicciones para el sexo femenino vs el sexo masculino, teniendo una tendencia de clasificar con más precisión al sexo femenino en el grupo de bajos ingresos.

### Pregunta 3: Aplique el algoritmo EqOddsPostProcessing sobre las predicciones del modelo de la pregunta 1. Obtenga las predicciones modificadas y evalúelas. ¿Qué cambios hubo en las métricas de Fairness? (2 ptos). 

#### Indicaciones adicionales:

- Obtenga tanto las intancias de BinaryLabelDatasetMetric como de ClassificationMetric y calcule al menos una métrica de Fairness.


### Respuesta

```python
# Utilizamos EqOddsPostprocessing como post procesamiento para cambiar las predicciones, mejorando la tasa de verdaderos positivos
# y falsos positivos
from aif360.algorithms.postprocessing.eq_odds_postprocessing import EqOddsPostprocessing


roc = EqOddsPostprocessing(unprivileged_groups=unprivileged_group,
                            privileged_groups=privileged_group,
                            seed = seed) # Utilizamos la misma métrica

roc.fit(dt_test, dt_pred)
```

```python
# Obtenemos las predicciones y evaluamos bajo las mismas métricas de Fairness
df_pred_eopp = roc.predict(dt_pred)

metric_sex_pred_eopp_df = BinaryLabelDatasetMetric(df_pred_eopp,
                                  unprivileged_groups=unprivileged_group,
                                  privileged_groups=privileged_group)

pred_vs_test_classification_metric_eopp = ClassificationMetric(dt_test, df_pred_eopp,
                                                    unprivileged_groups=unprivileged_group,
                                                    privileged_groups=privileged_group)
```

```python
# Calculamos 'Statistical Parity Difference' bajo las predicciones modificados con EqOddsPostprocessing

# Para el dataset de test
print(f'Pred EdOddsPostProcessing Statistical Parity Difference (sex): {metric_sex_pred_eopp_df.mean_difference()}')
```

Vemos que efectivamente con el método de post procesamiento nos hemos acercado al valor 0 de la diferencia estadística, por lo que hemos disminuido el sesgo en contra del grupo no privilegiado (sexo femenino) en torno a las predicciones obtenidas.

```python
# Ahora comparamos el ratio de error entre los datos de test y las predicciones modificadas
print(f'Error rate difference (unprivileged error rate - privileged error rate)= {pred_vs_test_classification_metric_eopp.error_rate_difference()}\n')
```

Similar a la tendencia de arriba, hemos disminuido a la mitad la diferencia de tasas de error entre el grupo no privilegiado y el grupo privilegiado. Esto puede indicar que la modificación de las predicciones mejora la justicia de aquellas con respecto al sexo femenino, en el que las tasas de errores de tanto este grupo como el grupo privilegiado (sexo masculino) son más similares en este momento.

### Pregunta 4: Aplique el algoritmo de Reweighing sobre los datos de entrenamiento. Compare las métricas de Fairness para los datos de entrenamiento antes y después de aplicar Reweighing. Luego, entrene un modelo de Regresión Logística utilizando los datos de entrenamiento modificados y obtenga las métricas de Fairness de las predicciones de dicho modelo. ¿Cómo varió el accuracy del modelo respecto al original y las métricas de fairness? (2 ptos).

#### Indicaciones adicionales:

- Obtenga tanto las intancias de BinaryLabelDatasetMetric como de ClassificationMetric y calcule al menos una métrica de Fairness.

- Considere 'Black' como grupo no privilegiado para la raza, 'White' como grupo privilegiado para la raza.
- Considere 'Female' como grupo no privilegiado para el sexo, 'Male' como grupo privilegiado para el sexo.

- Nuevamente, elija raza o sexo para el cálculo de las métricas. Está invitado(a) a realizar el análisis para ambos atributos sensibles, aunque **no es obligatorio**.

### Respuesta

```python
# Ahora aplicaremos cambios en los datos de entrenamiento para disminuir el sesgo entre grupos privilegiados y no privilegiados

# Preparamos el dataset de entrenamiento
df_train = X_train.copy()
df_train[target] = y_train

dt_train = BinaryLabelDataset(df=df_train, **attributes_params,
                             favorable_label=1, unfavorable_label=0)
metric_sex_train_df = BinaryLabelDatasetMetric(dt_train,
                                            unprivileged_groups=unprivileged_group,
                                            privileged_groups=privileged_group)
```

```python
# Vemos los pesos iniciales para el dataset de entrenamiento. Notamos que todos están seteados en 1
for i in dt_train.instance_weights:
    if i != 1:
        print(i)
```

```python
# Instanciamos algoritmo de Reweighing
from aif360.algorithms.preprocessing import Reweighing


reweight = Reweighing(unprivileged_groups=unprivileged_group,
                privileged_groups=privileged_group)
```

```python
# Obtenemos los nuevos datos de entrenamiento con nuevos pesos
reweighted = reweight.fit_transform(dt_train)
reweighted_sex_train_df = BinaryLabelDatasetMetric(reweighted,
                                              unprivileged_groups=unprivileged_group,
                                              privileged_groups=privileged_group)
```

```python
# Vemos los pesos iniciales ahora cambiaron para el dataset de entrenamiento
for i in reweighted.instance_weights:
    if i != 1:
        print(i)
```

```python
# Comparamos la diferencia de 'Statistical Parity Difference' entre los datos de entrenamiento y los datos modificados
print(f'Training Statistical Parity Difference (sex): {metric_sex_train_df.mean_difference()}')
print(f'Reweighted Statistical Parity Difference (sex): {reweighted_sex_train_df.mean_difference()}')
```

Vemos una diferencia notable entre el dataset de entrenamiento con pesos iniciales en 1, versus con pesos preprocesados. Después del pre procesamiento, prácticamente no hay diferencia estadística entre ambos grupos (sexo femenino vs sexo masculino).

```python
# Ahora entrenamos una regresión logística con los datos de entrenamiento, teniendo los pesos modificados
lr_rw = LogisticRegressionCV(solver='liblinear', cv=10, random_state=seed)
lr_rw.fit(X_train, np.ravel(y_train), sample_weight=reweighted.instance_weights)
```

```python
# Calculamos el nuevo accuracy y comparamos con el anterior
y_pred_proba_rw = lr_rw.predict_proba(X_test)[:,1]
y_pred_rw = y_pred_proba_rw >= 0.5
accuracy_rw = accuracy_score(y_test, y_pred_rw)

print(f"Accuracy normal: {accuracy}")
print(f"Accuracy reweighted: {accuracy_rw}")
```

En este caso, no vemos cambios significativos en el accuracy. De hecho, se ve que ligeramente el accuracy con nuevos pesos es ligeramente menor. Ahora mediremos métricas de Fairness.

```python
# Repetimos el proceso de calcular el dataset según los grupos privilegiados y no privilegiados pero ahora con las
# predicciones desde los datos con nuevos pesos

df_pred_rw = X_test.copy()
df_pred_rw[target] = y_pred_rw


dt_pred_rw = BinaryLabelDataset(df=df_pred_rw, **attributes_params,
                             favorable_label=1, unfavorable_label=0)
metric_sex_pred_rw_df = BinaryLabelDatasetMetric(dt_pred_rw,
                                            unprivileged_groups=unprivileged_group,
                                            privileged_groups=privileged_group)
```

```python
# Comparamos la diferencia de 'Statistical Parity Difference' entre las predicciones originales y 
# las predicciones con el modelo entrenado bajo nuevos pesos

print(f'Pred Original Statistical Parity Difference (sex): {metric_sex_pred_df.mean_difference()}')
print(f'Pred Reweighted Statistical Parity Difference (sex): {metric_sex_pred_rw_df.mean_difference()}')
```

Es muy notable la diferencia nuevamente con el pre procesamiento de los pesos, para el caso de las predicciones. Podemos ver que repesar los atributos bajo un grupo no privilegiado genera predicciones con menor sesgo y mejor Fairness, dado que la métrica estadística es más cercana a 0.

```python
# Finalmente, comparamos el ratio de error entre los datos de tests y las predicciones con los nuevos pesos

test_vs_pred_rw_classification_metric = ClassificationMetric(dt_test, dt_pred_rw,
                                                    unprivileged_groups=unprivileged_group,
                                                    privileged_groups=privileged_group)

print(f'Error rate difference before (unprivileged error rate - privileged error rate)= {pred_vs_test_classification_metric.error_rate_difference()}')
print(f'Error rate difference now (unprivileged error rate - privileged error rate)= {test_vs_pred_rw_classification_metric.error_rate_difference()}\n')
```

Nuevamente, notamos que la tasa de errores entre el dataset de test y las nuevas predicciones se acerca más a 0. Anteriormente estábamos ante una diferencia de errores en `-0.12` y ahora mejoramos a `-0.10`. Lo anterior también permite validar que en las predicciones dado el pre procesamiento de pesos se disminuye el sesgo de los datos. Importante considerar que esto es consistente con lo que ocurre en datos de entrenamiento: cuando los datos que ingresan a un modelo tienen un cierto sesgo, muy probablemente las predicciones tenderán a repetir este patrón de igual manera. Dado lo anterior, vale la pena aplicar pre y post procesamiento en torno a Fairness, ya que pese a que el `accuracy` puede mantenerse similar, el sesgo puede disminuir correctamente, haciendo un mejor uso de elementos de IA que sean más beneficiosos para la sociedad.
