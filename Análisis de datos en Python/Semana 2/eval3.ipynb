```python
apoyo8 = Info_sheet[Info_sheet["Día de llegada"] >= 414]
apoyo8 = apoyo8[['GRD', 'Número de llegada']].groupby(['GRD']).count()

apoyo8
```

```python
maxima_llegada = apoyo8['Número de llegada'].max()
total_pacientes_11_dias = apoyo8['Número de llegada'].sum()

print(f'Maxima llegada de pacientes por dia: {maxima_llegada}')
print(f'Total pacientes llegados en ultimos 11 dias: {total_pacientes_11_dias}')

print(f"Porcentaje de pacientes: {round((maxima_llegada / total_pacientes_11_dias) * 100, 2)}")
```
